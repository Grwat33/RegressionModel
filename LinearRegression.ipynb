```python
# import pandas
import pandas as pd
import numpy as np
```

```python
# import dataset
cardata = 'Dataset/cardata.csv'
cardata_df = pd.read_csv(cardata)
cardata_df
```

```python
# drop rows with empty values and duplicates
cardata_df = cardata_df.dropna(how = 'any')
cardata_df.drop_duplicates()
cardata_df
```

```python
# create manufacturer column and region (later) and keep popular manufacturers with more than 10 observations
cardata_df['manufacturer'] = cardata_df['name'].str.split(' ').str[0]
cardata_df['manufacturer'].unique()
manufacturers=cardata_df['manufacturer']
manufacturercount = cardata_df['manufacturer'].value_counts()
manufacturercount
manufacturercount.loc[lambda x : x>10]
cardata_df = cardata_df[cardata_df['manufacturer'].str.contains("Honda|Toyota|Ford|Chevrolet|Volkswagen|BMW|Nissan|Jaguar|Mercedes-Benz|Fiat|Audi|Lexus|Jeep|Mitsubishi")==True]
cardata_df['manufacturer'].unique()
cardata_df['manufacturerregion'] = cardata_df['name'].str.split(' ').str[0]
cardata_df['manufacturerregion'].unique()
manufacturers=cardata_df['manufacturerregion']
manufacturercount = cardata_df['manufacturerregion'].value_counts()
manufacturercount
manufacturercount.loc[lambda x : x>10]
cardata_df = cardata_df[cardata_df['manufacturerregion'].str.contains("Honda|Toyota|Ford|Chevrolet|Volkswagen|BMW|Nissan|Jaguar|Mercedes-Benz|Fiat|Audi|Lexus|Jeep|Mitsubishi")==True]
cardata_df['manufacturerregion'].unique()
```

```python
# create miles_per_gallon with mileage converted to mpg, drop outliers, view summary statistics
cardata_df['miles_per_gallon'] = cardata_df['mileage'].str.split(' ').str[0]
mileagemeasure = cardata_df['mileage'].str.split(' ').str[1].unique()
mileagemeasure
cardata_df['miles_per_gallon']=np.where(cardata_df.mileage.str.contains('km/kg'),cardata_df['miles_per_gallon'].astype(float)*0.0016747,cardata_df['miles_per_gallon'])
cardata_df['miles_per_gallon']=np.where(cardata_df.mileage.str.contains('kmpl'),cardata_df['miles_per_gallon'].astype(float)*2.3521458,cardata_df['miles_per_gallon'])
cardata_df['miles_per_gallon'] = cardata_df['miles_per_gallon'].astype('float')
from scipy import stats
cardata_df = cardata_df[(np.abs(stats.zscore(cardata_df['miles_per_gallon'])) < 3)]
cardata_df['miles_per_gallon'].describe()
```

```python
# create new maxpower column from max_power without bhp label, drop outliers
cardata_df['maxpower'] = cardata_df['max_power'].str.split(' ').str[0]
powermeasure = cardata_df['max_power'].str.split(' ').str[1].unique()
powermeasure
cardata_df['maxpower'] = cardata_df['maxpower'].astype('float')
cardata_df = cardata_df[(np.abs(stats.zscore(cardata_df['maxpower'])) < 3)]
cardata_df['maxpower'].describe()
```

```python
# create new engine_ column from engine without CC label, drop outliers
cardata_df['engine_'] = cardata_df['engine'].str.split(' ').str[0]
powermeasure = cardata_df['engine'].str.split(' ').str[1].unique()
powermeasure
cardata_df['engine_'] = cardata_df['engine_'].astype(float)
cardata_df = cardata_df[(np.abs(stats.zscore(cardata_df['engine_'])) < 3)]
cardata_df['engine_'].describe()
cardata_df
```

```python
# convert km_driven to miles_driven
cardata_df['miles_driven']=cardata_df.km_driven.astype(float)*0.621371
```

```python
# drop duplicate columns, change number formats
cardata_df = cardata_df[['name', 'manufacturer', 'manufacturerregion', 'year','selling_price','miles_driven', 'fuel', 'seller_type', 'transmission', 'owner', 'seats', 'miles_per_gallon', 'maxpower', 'engine_']]
cardata_df['seats'] = cardata_df['seats'].astype(int)
cardata_df['engine_'] = cardata_df['engine_'].astype(int)
cardata_df['miles_per_gallon'] = cardata_df['miles_per_gallon'].round(2)
cardata_df['owner'].value_counts()
```

```python
# reduce dataset to individual and dealer sales
cardata_df = cardata_df[cardata_df['seller_type'].str.contains("Trustmark Dealer")==False]
```

```python
# create new manufacturer region column
manufacturerregion = {'Honda' : 'Japan', 'Toyota' : 'Japan', 'Ford' : 'United States', 'Chevrolet' : 'United States', 'Jeep' : 'United States', 'Mercedes-Benz' : 'Europe',
       'Mitsubishi' : 'Japan', 'Audi' : 'Europe', 'Volkswagen' : 'Europe', 'BMW' : 'Europe', 'Nissan' : 'Japan', 'Lexus' : 'Europe', 'Jaguar' : 'Europe', 'Fiat' : 'Europe'}
cardata_df.manufacturerregion = [manufacturerregion[item] for item in cardata_df.manufacturerregion]
cardata_df
```

```python
# assign dummy variables
num_assign = {"manufacturer": {'Honda' : 1, 'Toyota' : 2, 'Ford' : 3, 'Chevrolet' : 4, 'Jeep' : 5, 'Mercedes-Benz' : 6,
                   'Mitsubishi' : 7, 'Audi' : 8, 'Volkswagen' : 9, 'BMW' : 10, 'Nissan' : 11, 'Lexus' : 12, 'Jaguar' : 13, 'Fiat' : 14},
                "manufacturerregion": {"Japan": 1, "United States": 2, "Europe": 3},
                "fuel": {"Petrol": 0, "Diesel": 1},
                "seller_type":  {"Individual": 0, "Dealer": 1},
                "transmission":     {"Automatic": 0, "Manual": 1},
                "owner":     {"Test Drive Car": 0, "First Owner": 1, "Second Owner": 2, "Third Owner": 3, "Fourth & Above Owner": 4},}
cardata_df = cardata_df.replace(num_assign)
cardata_df.head()
```

```python
# rename columns
cardata_df = cardata_df.rename(columns={"manufacturer":"Manufacturer", "manufacturerregion":"Region", "year":"Year", "selling_price":"SellingPrice($)", "miles_driven":"MilesDriven", "fuel":"FuelType", "seller_type":"SellerType", "transmission":"TransmissionType", "owner":"PrevOwnerCount", "seats":"SeatCount", "miles_per_gallon":"MPG", "maxpower":"Power(bhp)", "engine_":"EnginePower(CC)"})
```

```python
# drop and reorder columns
cardata_df = cardata_df[['Year', 'Manufacturer', 'Region', 'MilesDriven','FuelType','SellerType', 'TransmissionType', 'PrevOwnerCount', 'SeatCount', 'MPG', 'Power(bhp)', 'EnginePower(CC)', 'SellingPrice($)']]
```
